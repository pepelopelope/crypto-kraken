# Proyecto Final

Este año, la nota final se evaluará teniendo en cuenta un pequeño proyecto.
Este proyecto se realizará por parejas. Estas parejas no podrán haber estudiado juntos o estar en el mismo equipo a nivel laboral.


El proyecto consistirá en ser capaces de bajar la cotización de un par de monedas, un ejemplo de cotizaciones en Kraken:
https://www.kraken.com/es-es/prices?quote=EUR

Después de bajarlas, será necesario hacer una gráfica del movimiento que ha tenido.

Además, se hará una gráfica del oscilador estocástico:

https://www.novatostradingclub.com/indicadores/estocastico/
https://www.estrategiasdeinversion.com/herramientas/diccionario/analisis-tecnico/oscilador-estocastico-t-506#:~:text=La%20f%C3%B3rmula%20b%C3%A1sica%20para%20la,(Max%20%2D%20Mi)%20x%20100


La entrega final consistirá como mínimo en:

- Una memoria, la memoria será el contenido principal de la entrega. Y estará en PDF.
- Código, excluida cualquier librería adicional. Este código será material auxiliar para la corrección pero será obligado que también esté presente en la entrega.
- Estos dos elementos deberán estar contenidos dentro de un mismo zip. El zip deberá llamarse con los dos apellidos del realizador, si fueran varios, los dos apellidos de cada uno separados por guiones.
- Todo lo que no esté dentro de ese zip no se considerará (por ejemplo, no se puede decir que hay cierta información en una dirección de drive)
- No podrá tener un tamaño mayor de 10 MB.
- La memoria deberá tener los trozos de código a evaluar.
